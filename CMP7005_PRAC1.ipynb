```python
import pandas as pd

dongsi = pd.read_csv("CMP7005/PRSA_Data_Dongsi_20130301-20170228.csv")
guanyuan = pd.read_csv("CMP7005/PRSA_Data_Guanyuan_20130301-20170228.csv")
changping = pd.read_csv("CMP7005/PRSA_Data_Changping_20130301-20170228.csv")
huairou = pd.read_csv("CMP7005/PRSA_Data_Huairou_20130301-20170228.csv")
```

```python
!ls CMP7005
```

```python
dongsi['station'] = 'Dongsi'
guanyuan['station'] = 'Guanyuan'
changping['station'] = 'Changping'
huairou['station'] = 'Huairou'
```

```python
df = pd.concat([dongsi, guanyuan, changping, huairou], ignore_index=True)
```

```python
df['datetime'] = pd.to_datetime(df[['year', 'month', 'day', 'hour']])
```

```python
df = df.sort_values(by='datetime')
```
